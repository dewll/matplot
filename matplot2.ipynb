```python
import matplotlib.pyplot as plt
import numpy as np
fig =  plt.figure()

y = np.array([ 35, 25, 25, 15])
mylabels = ["Math","English","Physics","Chemistry"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,3, figsize=(16,5))

axes[0].set_title("Bar")
axes[0].grid()
axes[0].bar(n, n**2, align="center")


x = np.random.randn(100000)
axes[1].set_title("Histogram")
axes[1].hist(x)


y = np.array([5,6,7,8,9,2,1,3,4])
axes[2].set_title("Scatter")
axes[2].scatter(y, y+0.25*np.random.randn(len(y)))
```
