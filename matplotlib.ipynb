#MATPLOTLIB Tutorial

```python
from matplotlib import pylab
print(pylab.__version)
```

```python
import numpy as np
x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
```

```python
pylab.plot(x,y, 'r')
```

```python
pylab.subplot(1,3,1)
pylab.plot(x,y, 'r--')
pylab.subplot(1,3,2)
pylab.plot(x,y, 'b--')
pylab.subplot(1,3,3)
pylab.plot(x,y, 'g--')
```
